# Importing the dataset (Context data of students at 8th grade in Ireland)


```python
org_data = pd.read_spss("./data/orig/SPSS/bsgirlm8.sav")
org_data_copy = org_data.copy()
```

# Construction of Dataset

## Setting up  a new focused data set

### Adding Basic contextual variables(Gender, Language, Books at Home, Computers at Home, Shared_Computers, Background of Parents etc) - Categorical



```python

columns = [ "BSBG01", "BSBG03", "BSBG04",
    "BSBG05A", "BSBG05B", "BSBG05C", "BSBG05D",
    "BSBG05E", "BSBG05F", "BSBG06A",
    "BSBG06B", "BSBG07",
    "BSBG08A", "BSBG08B"]
df = org_data_copy[columns]
rename_dict = {
    "BSBG01": "Gender_Student",
    "BSBG03": "Lang_Spoken_Home",
    "BSBG04": "Books_Home",
    "BSBG05A": "Own_Computer",
    "BSBG05B": "Shared_Computer",
    "BSBG05C": "Smartphone",
    "BSBG05D": "Internet_Access",
    "BSBG05E": "Study_Desk",
    "BSBG05F": "Own_Room",
    "BSBG06A": "ParentA_Edu_Level",
    "BSBG06B": "ParentB_Edu_Level",
    "BSBG07": "Education_Expectation",
    "BSBG08A": "ParentA_Born_Country",
    "BSBG08B": "ParentB_Born_Country"
}
df = df.rename(columns=rename_dict)
df.head()
```

### Adding Scale variables of Plausible Values for each subject

```python
scale_vars = ["BSMMAT01", "BSMMAT02", "BSMMAT03", "BSMMAT04", "BSMMAT05",
    "BSSSCI01", "BSSSCI02", "BSSSCI03", "BSSSCI04", "BSSSCI05",
    "BSSBIO01", "BSSBIO02", "BSSBIO03", "BSSBIO04", "BSSBIO05",
    "BSSCHE01", "BSSCHE02", "BSSCHE03", "BSSCHE04", "BSSCHE05",
    "BSSPHY01", "BSSPHY02", "BSSPHY03", "BSSPHY04", "BSSPHY05",
    "BSSEAR01", "BSSEAR02", "BSSEAR03", "BSSEAR04", "BSSEAR05"
]
scale_var_dict = {
    "BSMMAT01": "1ST_PV_MATHEMATICS",
    "BSMMAT02": "2ND_PV_MATHEMATICS",
    "BSMMAT03": "3RD_PV_MATHEMATICS",
    "BSMMAT04": "4TH_PV_MATHEMATICS",
    "BSMMAT05": "5TH_PV_MATHEMATICS",
    "BSSSCI01": "1ST_PV_SCIENCE",
    "BSSSCI02": "2ND_PV_SCIENCE",
    "BSSSCI03": "3RD_PV_SCIENCE",
    "BSSSCI04": "4TH_PV_SCIENCE",
    "BSSSCI05": "5TH_PV_SCIENCE",
    "BSSBIO01": "1ST_PV_BIOLOGY",
    "BSSBIO02": "2ND_PV_BIOLOGY",
    "BSSBIO03": "3RD_PV_BIOLOGY",
    "BSSBIO04": "4TH_PV_BIOLOGY",
    "BSSBIO05": "5TH_PV_BIOLOGY",
    "BSSCHE01": "1ST_PV_CHEMISTRY",
    "BSSCHE02": "2ND_PV_CHEMISTRY",
    "BSSCHE03": "3RD_PV_CHEMISTRY",
    "BSSCHE04": "4TH_PV_CHEMISTRY",
    "BSSCHE05": "5TH_PV_CHEMISTRY",
    "BSSPHY01": "1ST_PV_PHYSICS",
    "BSSPHY02": "2ND_PV PHYSICS",
    "BSSPHY03": "3RD_PV_PHYSICS",
    "BSSPHY04": "4TH_PV_PHYSICS",
    "BSSPHY05": "5TH_PV_PHYSICS",
    "BSSEAR01": "1ST_PV_EARTH_SCIENCE",
    "BSSEAR02": "2ND_PV_EARTH_SCIENCE",
    "BSSEAR03": "3RD_PV_EARTH_SCIENCE",
    "BSSEAR04": "4TH_PV_EARTH_SCIENCE",
    "BSSEAR05": "5TH_PV_EARTH_SCIENCE"
}

scale_variables = org_data_copy[scale_vars].astype("float64").rename(columns=scale_var_dict)
df = pd.concat([df, scale_variables], axis=1)
```


### Adding remaining categorical variables - (Behavioural attributes including, dissorderly behaviour, bullying, likeness for subjects, confidence etc)

```python
behav_attr_variables = [
    "BSDGSSB", "BSDGSB", "BSDGSLM", "BSDGICM", "BSDGDML", "BSDGSCM", "BSDGSVM",
    "BSDGSLS", "BSDGICS", "BSDGSCS", "BSDGSVS", "BSDGSLB", "BSDGICB", "BSDGSCB",
    "BSDGSLE", "BSDGICE", "BSDGSCE", "BSDGSLC", "BSDGICC", "BSDGSCC", "BSDGSLP",
    "BSDGICP", "BSDGSCP", "BSDGSEC", "BSDGDSL", "BSDGDBL", "BSDGDEL", "BSDGDCL",
    "BSDGDPL", "BSDGVEP"
]
behav_attr = {
    "BSDGSSB": "Sense_of_School_Belonging",
    "BSDGSB": "Student_Bullying",
    "BSDGSLM": "Students_Like_Learning_Mathematics",
    "BSDGICM": "Instructional_Clarity_in_Mathematics_Lessons",
    "BSDGDML": "Disorderly_Behavior_during_Math_Lessons",
    "BSDGSCM": "Students_Confident_in_Mathematics",
    "BSDGSVM": "Students_Value_Mathematics",
    "BSDGSLS": "Students_Like_Learning_Science",
    "BSDGICS": "Instructional_Clarity_in_Science_Lessons",
    "BSDGSCS": "Students_Confident_in_Science",
    "BSDGSVS": "Students_Value_Science",
    "BSDGSLB": "Students_Like_Learning_Biology",
    "BSDGICB": "Instructional_Clarity_in_Biology_Lessons",
    "BSDGSCB": "Students_Confident_in_Biology",
    "BSDGSLE": "Students_Like_Learning_Earth_Science",
    "BSDGICE": "Instructional_Clarity_in_Earth_Science_Lessons",
    "BSDGSCE": "Students_Confident_in_Earth_Science",
    "BSDGSLC": "Students_Like_Learning_Chemistry",
    "BSDGICC": "Instructional_Clarity_in_Chemistry_Lessons",
    "BSDGSCC": "Students_Confident_in_Chemistry",
    "BSDGSLP": "Students_Like_Learning_Physics",
    "BSDGICP": "Instructional_Clarity_in_Physics_Lessons",
    "BSDGSCP": "Students_Confident_in_Physics",
    "BSDGSEC": "Digital_Self_Efficacy",
    "BSDGDSL": "Disorderly_Behavior_during_Science_Lessons",
    "BSDGDBL": "Disorderly_Behavior_during_Biology_Lessons",
    "BSDGDEL": "Disorderly_Behavior_during_Earth_Science_Lessons",
    "BSDGDCL": "Disorderly_Behavior_during_Chemistry_Lessons",
    "BSDGDPL": "Disorderly_Behavior_during_Physics_Lessons",
    "BSDGVEP": "Students_Value_Environmental_Preservation"
}


behavioral_attributes = org_data_copy[behav_attr_variables]
behavioral_attributes = behavioral_attributes[behav_attr_variables].rename(columns=behav_attr)
df = pd.concat([df, behavioral_attributes], axis=1)
```

## Removing Emptyl Attributes
Some empty attributes are present in the datafeame, which needs to be dropped.



```python
df = df.dropna(axis=1, how='all')
df.info()
```

###  Managing null rows
Since scale attributes have no null rows, the null values for the categorical variables needed to be taken care of. In this analysis, null rows will be dropped


```python
df = df.dropna(axis=0)
df.info()
```
